Next we create three functions, that we will help us in the chore of updating the covid-19 data.

```python
def archive_data(file_name, file_path):
    '''
    This function archives the old data file by renaming it with a date suffix in the format yymmdd,
    and then it prints a confirmation that the old data was archived including the name of the archived file
    
    Args:
        file_name (str): name of covid-19 data file
        file_path (str): path to covid-19 data file

    Returns:
        this function does not return anything
    '''

    file_full_path = os.path.join(file_path, file_name)

    try:
        mtime = os.path.getmtime(file_full_path)
        last_modified_time = datetime.datetime.fromtimestamp(mtime) 
        last_data_date = last_modified_time - timedelta(days=1)
        date_extention = str(last_data_date.year)[2:].zfill(2) + str(last_data_date.month).zfill(2) + str(last_data_date.day).zfill(2)
        new_file_name = file_name[:-4] + date_extention + file_name[-4:]
        new_file_full_path = os.path.join(file_path, new_file_name)
        os.rename(file_full_path, new_file_full_path)
        print("- Old data file archived as: '{}'".format(new_file_name))
        print()
    except OSError:
        mtime = 0
```

```python
def download_data(file_name, file_path):
    '''
    This function programmatically downloads the new data file from kaggle to the 'data' folder
    in the root directory of the web app. The new data file is compressed into .zip format
    
    Args:
        file_name (str): name of covid-19 data file
        file_path (str): path to covid-19 data file

    Returns:
        this function does not return anything
    '''
    
    !kaggle datasets download -d sudalairajkumar/novel-corona-virus-2019-dataset -f $file_name -p $file_path --force
    
```

```python
def extract_zip_file(file_name, file_path):
    '''
    This function unzips the compressed new data file found in the 'data' folder
    in the root directory of the web app. Then the zip file is deleted.
    
    Args:
        file_name (str): name of covid-19 data file
        file_path (str): path to covid-19 data file

    Returns:
        this function does not return anything
    '''
    
    for i,file in enumerate(os.listdir(file_path)):
        # looping through the files in the data folder, we search for the one ending with ".zip"
        if file.endswith(".zip"):
            try:
                # we obtain the zip file name, its full path and unzip it in the same folder
                zip_file_name = os.path.basename(file)
                zip_full_path = os.path.join(file_path, zip_file_name)
                
                with zipfile.ZipFile(zip_full_path, 'r') as zip_ref:
                    zip_ref.extractall(file_path)
                
                print()
                print("- New file extracted from compressed file to data folder as: '{}'".format(file_name))
                print()
                os.remove(zip_full_path)
                print('- Zip file deleted post extraction.')
            except OSError:
                pass
```

We utilize the above three functions in our main function for updating the covid-19 data: `get_latest_data()`

```python
def get_latest_data(file_name, file_path):
    
    '''
    This function is our main function for updating the covid-10 data. It will utilize our
    three helper functions 'archive_data()', 'download_data()' and 'extract_zipfile()' for completing 
    a sequence of tasks:
    
    - Rename current file data/covid_19_data.csv by adding a yymmdd date extension to the file name 
    and thereby, in effect, archiving it.
    - Download new file, which is compressed into .zip format
    - Extract data file 'covid_19_data.csv' from zip-file
    - Delete zip-file
    
    For each task in the sequence a confirmation is printed.
    Finally, a confirmation is printed once the full sequence is successfully completed: 
    'COMPLETED sequence for updating data successfully!'
    
    Args:
        file_name (str): name of covid-19 data file
        file_path (str): path to covid-19 data file

    Returns:
        this function does not return anything
    '''
    
    file_full_path = os.path.join(file_path, file_name)

    try:
        # get the time of last modification of the specified path; the getmtime method returns 
        # a floating point value which represents the number of seconds since the epoch. 
        # This method raise OSError if the file does not exist or is somehow inaccessible.
        mtime = os.path.getmtime(file_full_path)
        
        # converting floating point value to a datetime date
        last_modified_date = datetime.datetime.fromtimestamp(mtime).date()
        
        # compare last_modified_date of current file with today's date and only initiate sequence 
        # if the current data file is indeed old, ie. it was last modified yesterday or earlier
        today_date = datetime.date.today()
        
        if last_modified_date == today_date:
            print("We already have the latest data")
            
        elif last_modified_date < today_date:
            print('STARTING sequence for updating data...')
            print()
            
            archive_data(file_name, file_path)
            
            download_data(file_name, file_path)
            
            extract_zip_file(file_name, file_path)
            
            print()
            print('COMPLETED sequence for updating data successfully!')
        else:
            print('Error: Not able to comprehend; last_modified_date set to the future')
        
    except OSError:
        print('Error: Not able to extract meta data for provided file_name/file_path')
        
```

```python
def get_popdata(popdata_path = POPDATA_PATH):
    '''
    This function reads the population data from csv file into a pandas DataFrame, cleans up the
    column names and transforms 'Urban population ratio' data to type Float. 
    
    The data contains the following on country-level:
    
    - Country name
    - ISO code as described in the ISO 3166 international standard, the 3-letter version ('Short') 
    - Continent
    - Population
    - Population per square kilometer
    - Median age
    - Urban population ratio
    
    Args
        popdata_path (str): path to the population data file

    Returns:
        df_pop (dataframe): population dataframe
    '''
    
    # population dataset read into dataframe
    df_pop = pd.read_csv(popdata_path, delimiter=";") 
    
    # setting column headers
    df_pop.columns = ['Country', 'ISO', 'Continent', 'Population', 'Pop_km2', 'Median_age', 'Urban_Population_percent']
    
    # extracting the ratio from string column 'UrbanPopPercent' and replacing it with a new column with
    # float values for ratio.
    df_pop['Urban_Population_ratio'] = df_pop['Urban_Population_percent'].str.strip('%').astype(float).div(100)
    df_pop.drop('Urban_Population_percent', axis = 1, inplace=True)
    
    return df_pop
```

```python
def get_covid_data(data_path = DATA_PATH):
    '''
    This function reads the covid-19 data from csv file into a pandas DataFrame, cleans up the
    column names and updates some of the country names to match the ones used in the population dataset. 
    
    The data contains the following on country-level:
    - Country name
    - Date (Daily. The start date is not the same for all countries; it ranges from January to March 2020)
    - Deaths (total number for respective country at respective date)
    
    Args:
        data_path (str): path to covid-19 data file

    Returns:
        this function does not return anything
    '''
    
    df = pd.read_csv(data_path) # population dataset read into dataframe
    
    # clean data
    
    df = df.rename(columns={"ObservationDate": "Date", "Country/Region": "Country"})
    
    df = df.groupby(['Country', 'Date']).sum().reset_index()
    
    # update some of the country names to match the ones used in the population dataset
    
    df['Country'] = df['Country'].replace({'UK': 'United Kingdom', 
                                           'US': 'United States', 
                                           'Mainland China': 'China',
                                           'Czech Republic': 'Czechia',
                                           'Burma': 'Myanmar'
                                           })
    try:
        df['Country'] = df['Country'].replace({"('St. Martin',)": "St. Martin",
                                               "occupied Palestinian territory": "Palestine",
                                               "State of Palestine": "Palestine",
                                               "Ivory Coast": "Cote d'Ivoire",
                                               "Gambia, The": "Gambia",
                                               "The Gambia": "Gambia",
                                               "The Bahamas": "Bahamas",
                                               "Bahamas, The": "Bahamas",
                                               "Republic of Ireland": "Ireland",
                                               "Republic of the Congo": "Congo (Brazzaville)",
                                               " Azerbaijan": "Azerbaijan"
                                              
                                              })
    except:
        pass
        
    # sort on Country, then Date. Mostly it is only one row getting grouped, whereever data is on regional
    # groupby ensures we get the data summed up to country level
    df_covid = df.groupby(['Country', 'Date']).sum().reset_index()
    
    # dropping columns we will not use.
    df_covid.drop(['SNo', 'Confirmed', 'Recovered'], axis=1, inplace=True)
    
    return df_covid
```

```python
def merge_data():
    '''
    This function unzips the compressed new data file found in the 'data' folder
    in the root directory of the web app. Then the zip file is deleted.
    
    Args:
        file_name (str): name of covid-19 data file
        file_path (str): path to covid-19 data file

    Returns:
        this function does not return anything
    '''
    
    df_pop = get_popdata()
    
    df_covid = get_covid_data()
    
    df_merged = df_covid.set_index('Country').join(df_pop.set_index('Country')).reset_index()
    
    df_merged.dropna(subset=['Population'], inplace=True)

    df_merged['Deaths_per_100k'] = 100000*df_merged['Deaths']/(df_merged['Population'] + 1.0)
    
    df_merged = df_merged[['Continent', 'Country', 'ISO', 'Population', 'Urban_Population_ratio', 'Pop_km2', 'Median_age','Deaths', 'Deaths_per_100k', 'Date']]
    
    return df_merged
```

```python
def filter_data(df, historical=False):
    '''
    This function unzips the compressed new data file found in the 'data' folder
    in the root directory of the web app. Then the zip file is deleted.
    
    Args:
        file_name (str): name of covid-19 data file
        file_path (str): path to covid-19 data file

    Returns:
        this function does not return anything
    '''
            
    if historical:
        
        df_filtered = df.copy()
        df_filtered['New_deaths'] = 0.0
        df_filtered['New_deaths_smooth'] = 0.0
        
        country_list = df_filtered.Country.unique().tolist()
        
        for country in country_list:
            
        
            df_temp = df_filtered.loc[df_filtered.Country == country]
        
            df_temp['Shift1'] = df_temp.Deaths.shift(1).fillna(0)
            df_temp['New_deaths'] = df_temp['Deaths'] - df_temp['Shift1']
        
            df_temp['New_deaths_shift1'] = df_temp['New_deaths'].shift(1).fillna(0)
            df_temp['New_deaths_shift2'] = df_temp['New_deaths'].shift(2).fillna(0)
        
            df_temp['New_deaths_smooth'] = ((df_temp['New_deaths'] + df_temp['New_deaths_shift1'] + df_temp['New_deaths_shift2'])/3.0).astype(int)
        
            df_temp.drop(['Shift1', 'New_deaths_shift1', 'New_deaths_shift2'], axis=1, inplace=True)

            index_list = df_temp.index

            for i in index_list:
                df_filtered.loc[i,'New_deaths'] = df_temp.loc[i,'New_deaths']
                df_filtered.loc[i,'New_deaths_smooth'] = df_temp.loc[i,'New_deaths_smooth']
            
            
        df_filtered = df_filtered.reset_index(drop=True).set_index('Date')
        df_filtered.index = pd.to_datetime(df_filtered.index, infer_datetime_format=True)
        
    else:
        df_filtered = df[df.Date == df.Date.max()].reset_index(drop=True)
        
            
    return df_filtered
```

```python
def select_continent(df, continent):
    '''
    This function filters the dataframe on a single continent
    
    Args:
        df (dataframe): Dataset including a column 'Continent'
        continent (string): Name of continent. Valid values are found in below list

    Returns:
        df_continent (dataframe): Dataframe with only data from a single contintent
    '''
    if continent:
        
        continent_list = ['America', 'Europe', 'Asia', 'Africa', 'Oceania']
        
        assert continent in continent_list, "Continent is not in continent_list"
            
        df_continent = df[df['Continent'] == continent]
        
    return df_continent
```

```python
def rank_data(df_current, var, top_n = None):
    '''
    This function filters on the top n countries ranked on any of the following:
    
    - Population
    - Pop_km2
    - Urban_Population_ratio
    - Median_age
    - Deaths
    - Deaths_per_100k
    
    Args:
        file_name (str): name of covid-19 data file
        file_path (str): path to covid-19 data file

    Returns:
        this function does not return anything
    '''
    
    var_list = ['Population', 'Pop_km2', 'Urban_Population_ratio', 'Median_age', 'Deaths', 'Deaths_per_100k']
    assert var in var_list
    
    if top_n:
        tot_countries = df_current.shape[0]
        if top_n > tot_countries:
            top_n = tot_countries
            
    df = df_current.sort_values(by=[var], ascending=False)[:top_n].reset_index(drop=True)
    
    return df
```

```python
def prepare_barplot(continent=None, top_n = (None, None)):
    '''
    This funtion gets the current high level aggregated data suitable for bar plots.
    It is optional to filter on continent.
    It is also optional to filter on the top n countries ranked on any of the following:
    
    - Population
    - Pop_km2
    - Urban_Population_ratio
    - Median_age
    - Deaths
    - Deaths_per_100k
    
    Args:
        file_name (str): name of covid-19 data file
        file_path (str): path to covid-19 data file

    Returns:
        this function does not return anything
    '''

    df = merge_data()
    df_current = filter_data(df, historical=False)
    
    if continent:
        df_current = select_continent(df_current, continent)
        
    var_list = ['Population', 'Pop_km2', 'Urban_Population_ratio', 'Median_age', 'Deaths', 'Deaths_per_100k']
    
    # assert that top_n tuple input is valid in case two (no-None) elements are provided
    
    if (len(top_n) == 2) and (top_n[0] and top_n[1]):
        
        assert top_n[0] in var_list, "First tuple element in 'top_n' must be in var_list and of type String"
        assert isinstance(top_n[1], int) == True, "Second tuple element in 'top_n' must of type Integer"
    
        df_current = rank_data(df_current, top_n[0], top_n[1])
    
    return df_current
```

Playing around with automatically generated titles based on choice of continent and top_n in as input.

Earlier, when infection and recovery number were included, the bar chart could take on multiple variables. It still can, however, none of the var_lists have variables with similar magnitude of values. 

**Recommended: Some nice Choropleth map plots further down**

```python
def plot_bar(continent=None, top_n = ('Deaths_per_100k', 10)):
    
    
    var = top_n[0]
    n = top_n[1]
    
    df = prepare_barplot(continent=continent, top_n = top_n)
    
    df.set_index('Country', inplace=True)

    var_list_rel = ['Deaths_per_100k']
    keyword1 = '_per_100k'
    
    var_list_pop = ['Population', 'Pop_km2', 'Urban_Population_ratio']
    keyword2 = 'Pop'
    
    var_list_abs = ['Deaths']

    if keyword1 in var:
        var_list = var_list_rel
    
    elif keyword2 in var:
        var_list = var_list_pop
        
    else:
        var_list = var_list_abs
        
        
    tot_countries = df.shape[0]
    if n > tot_countries:
        top_n = tot_countries

    # continent related:
    # adjusting fig_with for continent Oceania which have very few countries 
    
    if continent == 'Oceania':
        fig_width = 5
    else:
        fig_width = 20
        
            
    if not continent:
        continent = "The World"
    
    df.sort_values(by=[var], ascending=False).loc[:, var_list][:n].plot.bar(
        title = '{} Top {}: Number of {}'.format(continent, n, var), 
        stacked = False,
        colormap='tab20b',
        figsize=(fig_width, 5)
        )
    
    plt.xticks(rotation=20)
    plt.legend(loc='upper right');
```

```python
def prepare_geo(df_current):
    
    fp = 'data/ne_10m_admin_0_countries.shp'

    gdf = gpd.read_file(fp)
    
    gdf = gdf.rename(columns={"ADMIN": "Country", "ADM0_A3": "Short"})
    
    gdf = gdf[['Country', 'Short', 'geometry']]
    gdf.columns = ['Country', 'ISO', 'geometry']
    
    
    df_all = df_current
    
    
    df = gdf.merge(df_all, how='left', on='ISO').drop(columns=['Date', 'Country_y'])
    
    df.rename(columns={"Country_x": "Country"}, inplace = True)
    
    df.fillna(0, inplace = True)
    
    return df
    
```

```python
def plot_geo(colname, continent=None):
    
    
    data = prepare_geo(df_current)
    
    data = data[data.Country!="Antarctica"]

    if continent:
        data = data[data.Continent == continent] 

    fig, ax = plt.subplots(1, 1, figsize=(16, 12))
    
    
    if not continent:
        #ax.set(xlim=(-24, 40), ylim=(35, 75))
        fontsize = 6
    elif continent == 'Europe':
        ax.set(xlim=(-24, 40), ylim=(35, 75))
        fontsize = 8
        ax.text(1, 46, 'France', fontsize=fontsize)
        ax.text(8, 62, 'Norway', fontsize=fontsize)
        
    elif continent == 'America':
        ax.set(xlim=(-130, -30), ylim=(-55, 75))
        fontsize = 10     
        ax.text(-103, 38, 'USA', fontsize=fontsize)    
        ax.text(-107, 53, 'Canada', fontsize=fontsize)
    elif continent == 'Asia':
        ax.set(xlim=(20, 145), ylim=(-10, 75))
        fontsize = 8     
    else:
        fontsize = 8

    

    data.plot(column=colname, ax=ax, legend=True, cmap='Wistia', legend_kwds={'label': "{}".format(colname), 
                                                                      'orientation': "vertical"})

    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    #plt.xticks([10.7, 10.8, 10.9])
    #plt.yticks([59.8, 59.9, 60.0])
    #plt.suptitle('Marka', y=0.73, fontsize=10)

    plt.title('{}: {}\n'.format(continent, colname),  fontsize=18)
    
    #plt.axis('off')

    skip_list = ['Norway', 'France', 'United States of America', 'Canada']

    for idx, row in data.iterrows():
        
        if (len(str(row['Country'])) > 3) & (str(row['Country']) not in skip_list):
            plt.annotate(s=row['Country'], xy=row['geometry'].centroid.coords[0],
                     horizontalalignment='center', fontsize=fontsize)
```
