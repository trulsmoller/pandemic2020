```python
def prepare_data(data, top_n_countries = -1, country_list = 'full'):
    '''
    Prepares any of the time series datasets for plotting as time series
    
    INPUTS
    data (DataFrame)- one of our time series datasets; confirmed, deaths, recovered
    top_n_countries (integer, optional) - based on the latest number of occurrences
    country_list (list of strings, optional) - if provided, top_n_countries will not be used
    
    OUTPUT
    df_clean (DataFrame) - prepared and clean dataframe
    '''
    
    drop_value = ['Cruise Ship', 'San Marino']
    data = data[~data['Country/Region'].isin(drop_value)]
    
    if country_list != 'full':
        top_n_countries = -1
    
    data_last_date = list(data.columns[-1:])[0]
    
    df_clean = data.groupby('Country/Region').sum().sort_values(by=[data_last_date], ascending=False)[:top_n_countries]
    
    df_clean.drop(['Lat', 'Long'], axis=1, inplace=True)
    
    
    
    df_clean = df_clean.T
    df_clean.index.names = ['Date']
    df_clean = df_clean.rename(columns={"Korea, South": "South Korea", "US": "United States"})
    df_clean.index = pd.to_datetime(df_clean.index)
    
    if country_list != 'full':
        df_clean = df_clean[country_list]
    
    print('Countries prepared for plotting:\n')
    print(list(df_clean.columns))
    
    return df_clean
```

```python
def makeplot(data, start_n_days_ago, country_list = 'full'):
    
    if country_list == 'full':
        country_list = list(data.columns)
    
    date_n_days_ago = str(pd.to_datetime((data.index + pd.DateOffset(-start_n_days_ago))[-1]).date())
    
    df = data.loc[date_n_days_ago:]

    axes = df[country_list].plot(marker='.', alpha=0.5, figsize=(10, 20), subplots=True)
    for ax in axes:
        ax.set_ylabel(data.name)
        ax.legend(loc='upper left')
```

```python
def prep_per_100k(data, pop_data):
    '''
    Prepares any of the time series datasets for plotting accumulated occurrences per 100k people
    at the most current date and sorting it descending.
    
    INPUT
    data (DataFrame)- one of our time series datasets; confirmed, deaths, recovered
    
    OUTPUT
    df (DataFrame) - prepared Dataframe
    '''
    
    # transposing data from the most current date, reindexing and changing column names
    df = data[-1:].T.reset_index()
    df.columns = ['Country', data.name]
    
    # initializing a new column Population  
    df['Population'] = 0
    
    # updating Population column from pop_data whenever there is a match on country name
    for i in range(df.shape[0]):
        for j in range(pop_data.shape[0]):
            a = df.Country[i]
            b = pop_data['Country (or dependency)'][j]
        
            if a == b:
                
                # column 3 (index 2) in df and column 2 (index 1) in pop_data 
                # are associated with population
                df.iat[i, 2] = pop_data.iat[j, 1]
                
    # adding column with the same name as 'data' with the suffix '_per_100k'
    colname = data.name + '_per_100k'
    
    df[colname] = 100000*df[data.name]/df.Population
    
    df = df.sort_values(by=[colname], ascending=False).reset_index()
    
    df.drop(['index'], axis=1, inplace=True)
    
    return df
```

```python
confirmed_per_100k = prep_per_100k(confirmed, pop)
confirmed_per_100k[:5]
```

```python
deaths_per_100k = prep_per_100k(deaths, pop)
deaths_per_100k[:5]
```

```python
recovered_per_100k = prep_per_100k(recovered, pop)
recovered_per_100k[:5]
```
